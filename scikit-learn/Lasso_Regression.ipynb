# Lasso Regression

- Lasso Regression is very similar to Ridge Regression
- But, there are some critical difference which slightly change the behaviour
- Overall Lasso Regression is also used for the reguilarization effect (decresing variance) as discussed in RidgeRegression

## How is Lasso Regression different from Ridge Regression?

- In Ridge Regression, we minimized the value of 
    (sum of sqaured residuals + lambda*slope^2)
- In Lasso Regression, we minimize the value of 
    (sum of sqaured residuals + lambda* |slope|)
    
Overall the resultant effect is similar to Ridge Regression. The same can be applied with a variety pf ata and different contexts to slightly reduce the variance by fitting a lesser slope curve.

### Main difference

- If lamdba = 0, we get the least squared formula.
- with increaing lambda value in case of Ridge Regression, we can get a line which is close to 0 slope
- But, with Lasso Regression we can actually get a line with a slope 0

Here lets asume, y is highly dependant on the value of x1 and not so much on x2

Eg: 
y = w0 + w1x1 + w2x2;  

by using Ridge Regression we will try to minimize:

    lambda*(w1^2 + w2^2)  ---> here theoritically we will not be able to reduce the value of w2 to 0

But by using Lasso Regression, we minimize:

    lambda*(|w1| + |w2|)  ---> here we can get a value of 0 for w2

## Where is Lasso Regression better? and, Where is Ridge Regression better?

- Lassso Regression is especially useful when we have a lot of unnecessary variables which wont affect the result much
- Ridge Regression is useful when a lot of our varibles contribute towards calculating the output

# Example

```python
from sklearn import linear_model
```

```python
reg = linear_model.Lasso(alpha = 0.05) # alpha == lambda
reg.fit([[0, 0], [1, 1]], [0, 1])
```

```python
reg.predict([[1, 1]])
```
